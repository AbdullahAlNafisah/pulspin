```python
from controller import GRBL

position = GRBL("COM15")
```

```python
# IMPORTANT ABOUT YOUR MOTOR WIRING:
# GRBL controls AXES by firmware mapping:
#   "X axis" is whatever driver is wired to X in GRBL (your Xmotor2)
#   "Y axis" is whatever driver is wired to Y in GRBL (your Ymotor2)
#
# If you connected another motor to Ymotor1, GRBL will NOT control it
# unless Ymotor1 is electrically tied to the same Y STEP/DIR signals
# as Ymotor2 (i.e., mirrored/parallel signals) OR the firmware is set up
# for a second Y driver (less common).
#
# This Python code sends X/Y G-code only. Extra motor behavior depends
# on how you wired that Ymotor1 driver signals.

# 1) Manually move the head to your real (0,0), then:
position.origin_here()

# 2) Move in millimeters:
position.move(dx=10, dy=0)      # 10mm east
position.move(dx=0, dy=20)      # 20mm north
print("pos:", position.x, position.y)
```
